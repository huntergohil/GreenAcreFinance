```python
# 🧠 Investment Intelligence PoC: Real Data Integration (Colab Version)## 📦 Install Required Packages!pip install yfinance requests## 📥 Importsimport yfinance as yfimport requestsimport numpy as npfrom sklearn.model_selection import train_test_splitfrom sklearn.ensemble import RandomForestRegressor## 🔌 1. Real Data Collection Functionsdef fetch_financial_news():    url = "https://newsapi.org/v2/everything"    params = {        "q": "Norwegian offshore drilling OR Seadrill OR Valaris",        "from": "2024-01-01",        "sortBy": "publishedAt",        "apiKey": "YOUR_NEWSAPI_KEY"  # Replace with your NewsAPI key    }    response = requests.get(url, params=params)    articles = response.json().get("articles", [])    sentiment = "positive" if len(articles) > 0 else "neutral"    return {        "energy_sector_sentiment": sentiment,        "key_events": [a['title'] for a in articles[:5]]    }def fetch_market_data():    eqnr = yf.Ticker("EQNR.OL")    dno = yf.Ticker("DNO.OL")    return {        "stock_prices": {            "major_offshore_companies": [                {"Equinor": eqnr.history(period="1d")['Close'].iloc[-1]},                {"DNO": dno.history(period="1d")['Close'].iloc[-1]}            ]        }    }def fetch_geopolitical_data():    return {        "tension_index": 0.7,        "key_regions": ["Middle East", "Russia-Ukraine border"]    }## 🧱 Data Ingestion Classclass DataIngestion:    def collect_data(self):        return {            "financial_news": fetch_financial_news(),            "market_data": fetch_market_data(),            "geopolitical_reports": fetch_geopolitical_data()        }## 📊 Predictive Model Classclass PredictiveModel:    def __init__(self, features, target):        self.features = features        self.target = target        self.model = None    def prepare_data(self):        X_train, X_test, y_train, y_test = train_test_split(            self.features, self.target, test_size=0.2, random_state=42        )        return X_train, X_test, y_train, y_test    def train_model(self):        X_train, X_test, y_train, y_test = self.prepare_data()        self.model = RandomForestRegressor(n_estimators=100, random_state=42)        self.model.fit(X_train, y_train)        return {            "train_score": self.model.score(X_train, y_train),            "test_score": self.model.score(X_test, y_test)        }    def predict(self, new_data):        return self.model.predict(new_data)## 🧠 Insight Generation Classclass InsightGenerator:    def __init__(self, data, predictions):        self.data = data        self.predictions = predictions    def generate_insights(self):        sentiment = self.data.get('financial_news', {}).get('energy_sector_sentiment', 'neutral')        geo_risk = self.data.get('geopolitical_reports', {}).get('tension_index', 0)        volatility = np.std(self.predictions)        recommendations = []        if sentiment == 'positive':            recommendations.append("Potential positive outlook for offshore drilling")        if geo_risk > 0.6:            recommendations.append("Caution advised due to high geopolitical tensions")        return {            "overall_sentiment": sentiment,            "risk_assessment": {                "geopolitical_risk": geo_risk,                "market_volatility": volatility            },            "key_recommendations": recommendations        }## 🚀 Run Everythingdef run_full_pipeline():    # Ingest    data = DataIngestion().collect_data()    # Predict    features = np.random.rand(100, 5)    target = np.random.rand(100)    model = PredictiveModel(features, target)    model.train_model()    prediction = model.predict(np.random.rand(1, 5))    # Insights    insights = InsightGenerator(data, prediction).generate_insights()    return {        "data_collected": data,        "prediction": prediction.tolist(),        "insights": insights    }results = run_full_pipeline()results
```
